```python
import numpy as np 
import pandas as pd 
```

```python
df =pd.read_csv('animes.csv')
print(df.info())
```

#cleaning for recommender


```python
#cleaning the genre column
df["genre"]=df["genre"].apply(eval)
#filling the mising data with '_'
#df.fillna('_',inplace=True)
df=df.drop_duplicates(subset=['uid'], keep='first')
```

```python
df.reset_index(drop=True,inplace=True)
print(df.shape)
df.head()
```

```python
#splitting the startdate into start and end dates
df[['Start','End']] = df['aired'].str.split('to',expand=True).replace(' ', '')
df['End'] = df['End'].str.strip()
#splitting the genre columns marking each genre in the genre column
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('genre')),
                          columns=mlb.classes_,
                          index=df.index))
#dropping the aired date column
df.drop(['aired'], axis=1,inplace=True)
df.to_csv('animes_cleaned.csv',index=False)
df.head()
```

```python
print(df.info())
```

#creating a new df for the individual anime pages

```python
#df =pd.read_csv('animes.csv')
df[['Start','End']] = df['aired'].str.split('to',expand=True).replace(' ', '')
df['End'] = df['End'].str.strip()
df.drop(['aired'], axis=1,inplace=True)
df["genre"]=df["genre"].apply(eval)
df['genre']=df['genre'].map(lambda x: ", ".join(x))
df=df.drop_duplicates(subset=['uid'], keep='first')
df.synopsis=df.synopsis.fillna("No synopsis has been added for this series yet. \r\n \r\n Click here  to update this information.")
df.synopsis=df.synopsis.replace(["No synopsis has been added for this series yet. \r\n \r\n Click here  to update this information."],"Sorry no summary available.")
df.episodes=df.episodes.astype(str)
df.score=df.score.astype(str)
df.episodes=df.episodes.fillna("Not available")
df.score=df.score.fillna("Not Available")
df.img_url=df.img_url.fillna("https://m.media-amazon.com/images/S/sash/4FyxwxECzL-U1J8.png")
#print(df.synopsis.value_counts())
print(df.info())
print(df.shape)
df.to_csv('animes_details_cleaned.csv',index=False)
```

```python
def Get_anime_UID(name):
  data = pd.read_csv('https://raw.githubusercontent.com/DibyaSadhukhan/Anime_recommender_data/main/Data/animes_cleaned.csv')
  uid=data.loc[data['title']==name]['uid'].values[0]
  del data
  return uid
def get_anime_recommendations(name):
  #get 10 recommendations for the searched anime.
  data = pd.read_csv('https://raw.githubusercontent.com/DibyaSadhukhan/Anime_recommender_data/main/Data/recommendations.csv')
  rec=data.loc[data['uid']==Get_anime_UID(name)]
  rec=rec.values[0]
  return rec  
def Get_recommendations_details(name):
  details=[]
  data=pd.read_csv('https://raw.githubusercontent.com/DibyaSadhukhan/Anime_recommender_data/main/Data/animes_details_cleaned.csv')
  rec=get_anime_recommendations(name)
  for i in rec[1:]:
    detail=data.loc[data['uid'] == i]
    details.append(dict(detail.iloc[0]))
  return details
Get_recommendations_details('Naruto')
```
