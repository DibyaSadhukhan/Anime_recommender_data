```python
import numpy as np 
import pandas as pd 
```

```python
df =pd.read_csv('animes.csv')
print(df.info())
```

```python
#cleaning the genre column
df["genre"]=df["genre"].apply(eval)
#filling the mising data with '_'
df.fillna('_',inplace=True)
```

```python
#splitting the startdate into start and end dates
df[['Start','End']] = df['aired'].str.split('to',expand=True).replace(' ', '')
df['End'] = df['End'].str.strip()
#splitting the genre columns marking each genre in the genre column
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('genre')),
                          columns=mlb.classes_,
                          index=df.index))
#dropping the aired date column
df.drop(['aired'], axis=1,inplace=True)
df.to_csv('animes_cleaned.csv',index=False)
df.head()
```
